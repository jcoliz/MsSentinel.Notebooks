# Hello World: Show me my tables

This quick example is a connectivity check to verify your ability to see tables from notebooks

## List Databases

Which databases are available in the system?

```python
from sentinel_lake.providers import MicrosoftSentinelProvider

sentinel_provider = MicrosoftSentinelProvider(spark)

sentinel_provider.list_databases()
```

## List System Tables

Which tables are available in the system tables database?

```python
tables = sentinel_provider.list_tables()

for table in tables:
    print(f"{table.databaseName:<15} | {table.name:<20}")
```

## List Workspace Tables

Which tables are available in one of my workspaces?

**IMPORTANT** Set `WORKSPACE_NAME` to one of the workspaces returned earlier in the "List Databases" cell

```python
# Configuration - Update this with your workspace name
WORKSPACE_NAME = "<YOUR_WORKSPACE_NAME>"

tables = sentinel_provider.list_tables(WORKSPACE_NAME)

for table in tables:
    print(f"{table.databaseName:<15} | {table.name:<20}")
```
